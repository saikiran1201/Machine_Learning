Introducing SciKit Learn

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
test_pred = model.predict(X_test)
test_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df["sales"].mean()
```

```python
sns.histplot(data=df, x="sales")
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(test_pred, y_test)
print('The rsquared value is: ' + str(round(r2, 3)))
```

```python
mean_absolute_error(y_test, test_pred)
```

```python
#10 percent error when compare with the avg
#msE is varience of predictions
mean_squared_error(y_test, test_pred)
```

```python
#RMSE is sd of pedictions
np.sqrt(mean_squared_error(y_test, test_pred))
```

```python
test_resuduals = y_test - test_pred
test_resuduals
```

```python
sns.scatterplot(x= y_test, y = test_resuduals)
#random
plt.axhline(y = 0, color = "r", ls="--")
```

```python
sns.displot(test_resuduals,bins=25 ,kde = "True")
```

Still unsure if normality is a reasonable approximation? We can check against the normal probability plot.

```python
import scipy as sp
# Create a figure and axis to plot on
fig, ax = plt.subplots(figsize=(6,8),dpi=100)
# probplot returns the raw values if needed
# we just want to see the plot, so we assign these values to _
_ = sp.stats.probplot(test_resuduals,plot=ax)
```
